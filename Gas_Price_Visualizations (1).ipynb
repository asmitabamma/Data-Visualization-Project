```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
Gas_df = pd.read_csv('Gas_Prices.csv')
Gas_df
```

```python
# Finding out columns 
Gas_df.columns
```

**Line** **Plot**

```python
# Creating a line plot

fig = plt.figure(figsize=(10,5))

ax1 = fig.add_axes([0, 0, 1, 1])

ax1.set_title('Gas Prices of USA, UK, Japan and Canada comparisons from (1990-2008)')

ax1.plot(Gas_df['Year'],
         Gas_df['USA'],
         color='green', label='USA',)

ax1.plot(Gas_df['Year'],
         Gas_df['UK'], 
         color='blue', label = 'UK')

ax1.plot(Gas_df['Year'],
         Gas_df['Japan'], 
         color='Orange', label = 'Japan')

ax1.plot(Gas_df['Year'],
         Gas_df['Canada'], 
         color='Red', label = 'Canada')
Gas_df.Year[::3]
plt.xticks(Gas_df.Year)

plt.xlabel('Year', size=13)
plt.ylabel('Dollars per gallon', size=13)
plt.legend()
plt.show
plt.savefig('Line_Plot.jpg', dpi=300, bbox_inches='tight')
```

**Bar Graph**

```python
# Creating a bar graph

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_axes([0, 0, 1, 1])
a = Gas_df.sort_values('Year')
b = Gas_df.sort_values('USA',)
ax1.set_title('Comparing Gas Prices of USA from 1990-2008')
ax1.bar(a.Year, b.USA, color='orange')
Gas_df.Year[::3]
plt.xticks(Gas_df.Year)


plt.xlabel('Year', size=13)
plt.ylabel('USA', size=13)
fig.savefig('Bar_Graph.jpg', dpi=300, bbox_inches='tight')
```

**Scatter** **Plot**

```python
# Creating a scatter plot

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.set_title('Gas Prices between two countries, USA vs UK (1990-2008)')
ax1.scatter(Gas_df.UK, Gas_df.USA, color='Red', s=100)

plt.xlabel('UK',size=13)
plt.ylabel('USA',size=13)
fig.savefig('ScatterPlot.jpg', dpi=300, bbox_inches='tight')
```
